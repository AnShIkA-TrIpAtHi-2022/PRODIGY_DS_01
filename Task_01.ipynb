```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./dataset_task_01.csv')
display(data)
```

Here, I have displayed three Histogram visualisations of the dataset. The dataset was taken from a survey done to analyse Volunteering among college students. The question analysed here inquires about the frequency of being a volunteer and an individual can choose any of the four options, namely : Sometimes, Rarely, Frequently, Never.

```python
plt.bar(data["COURSE"],data["NAME"],width=0.6)
plt.xlabel("Courses")
plt.ylabel("Frequency OF Students")
plt.yticks([])
plt.xticks(rotation = 90)
plt.title("Distribution of Student's courses")
plt.show()
```

The above bar chart or histogram shows carious frequencies of students who responded the survey based on the courses.

```python
plt.figure(figsize=(10, 6))
plt.bar(data['OPTION'], data['NAME'], width=0.6)
plt.xlabel('Responses')
plt.ylabel('Frequency')
plt.yticks([])
plt.title('Distribution of Student Responses')
plt.show()
```

The bar chart or histogram above shows the distribution of students based on their responses to the question.
